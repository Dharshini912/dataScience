```python
import matplotlib.pyplot as plt
import numpy as np
x=np.arange(0,11)
y=x
print(x)
print(y)
plt.figure(figsize=(6,4))
plt.plot(x,y,marker="X",linestyle="-",color="orange")
plt.title("Line plot")
plt.xlabel("X1 value")
plt.ylabel("X2 value")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
x = ['A', 'B', 'C', 'D']
y = [40, 70, 30, 90]
plt.figure(figsize=(6,4))
plt.bar(x, y, color="blue")
plt.title("Bar Plot")
plt.xlabel("Products")
plt.ylabel("Sales Values")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(11)
x = np.arange(1, 11)
y = np.random.randint(10, 100, size=10)
plt.figure(figsize=(8,6))
plt.subplot(2, 2, 1)
plt.plot(x, y, color="blue")
plt.title("Line Plot")
plt.subplot(2, 2, 2)
plt.bar(x, y, color="green")
plt.title("Bar Plot")
plt.subplot(2, 2, 3)
plt.scatter(x, y, color="red")
plt.title("Scatter Plot")
plt.subplot(2, 2, 4)
plt.hist(y, color="purple")
plt.title("Histogram")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
data = np.random.randn(200)
plt.hist(data, bins=20, edgecolor="black")
plt.title("Histogram of Normal Distribution")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```
